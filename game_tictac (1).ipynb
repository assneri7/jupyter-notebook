```python
#importing important module for guessing random generated number game
import random
import time
# generate a random number n from 1 to 99

n = random.randint(1,100)
print("Welcome to Number guessing game\nYou only have only 5 chance")
#delay 2 seccond after welcomeing message
time.sleep(2)
count = 5
guess = int(input("Enter your first guess: "))
#loop to repat the guess for 5 times else the player will lose the game
while guess != n and count > 1:
    if guess > n:
        print("Too High")
        count -=1
        print("you have only {} chances".format(count))
        guess = int(input("Enter your first guess: "))
    elif guess < n:
        print("Too Low")
        count -=1
        print("you have only {} chances".format(count))
        guess = int(input("Enter your first guess: "))
    else:
        break
if guess == n:
    print("Good Job The secrete number {}".format(n))
else:
    print("You lost the game The secrete number {}, Try again".format(n))
```
